```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_anime_by_id(target_anime_id, df_clean, x, threshold=0.8):
    # Check if the target anime ID exists in the dataframe
    if target_anime_id not in df_clean['anime_id'].values:
        raise ValueError(f"Anime ID {target_anime_id} not found in the dataset.")

    # Get the index of the target anime
    target_index = df_clean.index[df_clean['anime_id'] == target_anime_id][0]

    # Calculate cosine similarity
    similarity_matrix = cosine_similarity(x)
    similarity_scores = similarity_matrix[target_index]

    # Find indices of similar anime based on the threshold
    similar_indices = [i for i, score in enumerate(similarity_scores) if score > threshold and i != target_index]
    similar_indices = sorted(similar_indices, key=lambda i: similarity_scores[i], reverse=True)

    # Get anime IDs of the recommended anime using iloc
    recommended_anime_ids = df_clean.iloc[similar_indices]['anime_id'].values

    return recommended_anime_ids, similarity_scores[target_index]
```
